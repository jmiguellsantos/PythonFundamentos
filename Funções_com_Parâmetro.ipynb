```python
def soma(a,b):
  return a + b


  
```

```python
def multiplica(num1, num2):
  return num1 * num2
```

```python
def outra(num1, b, msg):
  return (num1 + b) * msg
```

```python
print(soma(2, 5))
```

```python
print(soma(10, 20))
```

```python
print(multiplica( 4, 5))
print(multiplica(2, 8))
```

```python
print(outra(3, 2, 'João'))
```

```python
#Nomeando Parâmetros
```

```python
def nome_completo(string1, string2):
  return f'Seu nome completo é {string1} {string2}'
```

```python
print(nome_completo('João Miguei', 'Oliveira dos Santos'))
```

```python
#A diferença entre Parâmetros e Argumentos
```

```python
#Parâmetros são variáveis declaradas na definição de uma função
```

```python
#Argumentos são dados passados durante a execução de uma função
```

```python
#A ordem dos parâmetros importam

nome = 'Felicidade'

sobrenome = 'jones'
```

```python
print(nome_completo(nome, sobrenome))
```

```python
#Argumentos nomeados (Keyword Arguments)
#Caso utilizemos nomes dos parâmetros nos argumentos para informá-los, podemos utilizar qualquer ordem
```

```python
def soma_impares(numeros):
  total = 0
  for num in numeros:
    if num % 2 != 0:
      total = total + num
  return total


```

```python
lista = [1, 2, 3, 4, 5, 6, 7]
print(soma_impares(lista))
```
